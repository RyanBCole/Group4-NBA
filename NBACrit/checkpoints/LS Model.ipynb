```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
from sklearn import preprocessing 

# Read the data
X_full = pd.read_csv(r"C:\Users\redye\Desktop\NBACrit\final_df.csv")
X_full["GAME_DATE_EST"] = pd.to_datetime(X_full["GAME_DATE_EST"],infer_datetime_format=True)
X_full = X_full.sort_values(by=["GAME_DATE_EST"], ascending=False)
X_full = X_full.reset_index()



# Label / One-Hot encoding team IDs and season
le = preprocessing.OneHotEncoder()
le2 = preprocessing.LabelEncoder()
ohe_home = le.fit_transform(X_full[["home_team"]]).toarray()
ohe_away = le.transform(X_full[["away_team"]]).toarray()

# Drop old columns
X_full = X_full.drop(columns =[
    "home_team","away_team", "PTS_home", "PTS_away","GAME_DATE_EST","index", "book_name_over", "book_name_under"], axis=1)
```

```python
y = X_full["PTS_comb_actual"]
X_full = X_full.drop(columns=['PTS_comb_actual'])

test = len(X_full) - int(len(X_full)*0.9)

X_train = X_full.iloc[test:]
X_valid = X_full.iloc[:test]
y_train = y.iloc[test:]
y_valid = y.iloc[:test]

```

```python
data = {
    "Actual_Total": y_valid,
    "ls_prediction": pred_ls
}
```

```python
predict_ls = pd.DataFrame(data)
predict_ls
```

```python
predict_ls['over_line'] = X_valid["total1_over"]
predict_ls['under_line'] = X_valid["total1_under"]
```

```python
predict_ls['over_difference'] = predict_ls['ls_prediction'] - predict_ls['over_line']
predict_ls['under_difference'] = predict_ls['under_line'] - predict_ls['ls_prediction']
```

```python
predict_ls
```

```python
def determine_bet(row):
    if row['over_difference'] >= 10:
        return 'Over'
    elif row['under_difference'] > 5:
        return 'Under'
    else:
        return 'No Bet'
    
predict_ls['Bet'] = predict_ls.apply(determine_bet, axis=1)
```

```python
predict_ls['over_odds'] = X_valid['over_odds']
predict_ls['under_odds'] = X_valid['under_odds']
predict_ls
```

```python
def convert_to_decimal(odds):
    return (odds / 100) + 1 if odds >= 100 else (100 / abs(odds) + 1)

# Apply the conversion function to the 'over' column and create a new 'decimal_over' column
predict_ls['over_odds'] = predict_ls['over_odds'].apply(convert_to_decimal)
predict_ls['under_odds'] = predict_ls['under_odds'].apply(convert_to_decimal)
predict_ls
```

```python
def determine_outcome(row):
    if row['Bet'] == 'Over':
        if row['over_line'] < row['Actual_Total']:
            return 'Win'
        elif row['over_line'] > row['Actual_Total']:
            return 'Loss'
        else:
            return 'Push'
    elif row['Bet'] == 'Under':
        if row['under_line'] < row['Actual_Total']:
            return 'Loss'
        elif row['under_line'] > row['Actual_Total']:
            return 'Win'
        else:
            return 'Push'
    else:
        return None

predict_ls['outcome'] = predict_ls.apply(determine_outcome, axis=1)
```

```python
predict_ls
```

```python
def calculate_profit(row):
    if row['outcome'] == 'Win':
        if row['Bet'] == 'Over':
            return row['over_odds'] - 1
        elif row['Bet'] == 'Under':
            return row['under_odds'] - 1
    elif row['outcome'] == 'Loss':
        return -1
    return None

# Apply the function to create the 'profit' column
predict_ls['profit'] = predict_ls.apply(calculate_profit, axis=1)
predict_ls
```

```python
predict_ls['game_id'] = X_valid['game_id']
```

```python
predict_ls.set_index('game_id', inplace=True)
predict_ls
```
