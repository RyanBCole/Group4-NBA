```python
import pandas as pd
```

```python
best_over = pd.read_csv(r'C:\Users\redye\Desktop\NBACrit\best_over.csv')
```

```python
best_over
```

```python
index_to_drop = best_over[best_over['total1'] == 141.5].index

# Drop the rows with the specified index
best_over = best_over.drop(index_to_drop)
best_over
```

```python
def convert_to_decimal(odds):
    return (odds / 100) + 1 if odds >= 100 else (100 / abs(odds) + 1)

# Apply the conversion function to the 'over' column and create a new 'decimal_over' column
best_over['decimal_over'] = best_over['over'].apply(convert_to_decimal)
best_over
```

```python
best_over = best_over.sort_values(by='decimal_over', ascending=False)
best_over
```

```python
first_values = best_over.groupby('game_id')['total1'].transform('first')
best_over['difference'] = (best_over['total1'] - first_values)*2

best_over
```

```python

# Function to calculate the compound factor based on the difference
def calculate_compound_factor(difference):
    return (1 + 0.022) ** difference

# Calculate the difference with respect to the first value of "total_1" for each "game_id"
first_values = best_over.groupby('game_id')['total1'].transform('first')

# Calculate the compound factor and create a new column "difference2"
best_over['difference2'] = calculate_compound_factor(best_over['difference'])
```

```python
#subtract d2 from over and make new column for comparing lines on equal footing
best_over['equal_lines'] = best_over['decimal_over'] / best_over['difference2']
best_over
```

```python
max_equal_lines = best_over.groupby('game_id')['equal_lines'].transform('max')

# Filter the DataFrame to keep only the rows with the maximum value in the "equal_lines" column for each "game_id"
overs = best_over[best_over['equal_lines'] == max_equal_lines]

overs
```

```python
overs = overs.sort_values(by='equal_lines', ascending=False)
overs
```

```python
columns_to_drop = ['difference', 'difference2', 'equal_lines', 'under', 'decimal_over']
overs = overs.drop(columns_to_drop, axis=1)
overs
```

```python
overs = overs.sort_values(by='game_id', ascending=False)
overs
```

```python
best_under = pd.read_csv(r'C:\Users\redye\Desktop\NBACrit\best_under.csv')
best_under
```

```python
index_to_drop = best_under[best_under['total1'] == 141.5].index

# Drop the rows with the specified index
best_under = best_under.drop(index_to_drop)
best_under
```

```python
def convert_to_decimal(odds):
    return (odds / 100) + 1 if odds >= 100 else (100 / abs(odds) + 1)

# Apply the conversion function to the 'over' column and create a new 'decimal_over' column
best_under['decimal_under'] = best_under['under'].apply(convert_to_decimal)
best_under
```

```python
first_values = best_under.groupby('game_id')['total1'].transform('first')
best_under['difference'] = (best_under['total1'] - first_values)*2

best_under
```

```python

# Function to calculate the compound factor based on the difference
def calculate_compound_factor(difference):
    return (1 + 0.022) ** difference

# Calculate the difference with respect to the first value of "total_1" for each "game_id"
first_values = best_under.groupby('game_id')['total1'].transform('first')

# Calculate the compound factor and create a new column "difference2"
best_under['difference2'] = calculate_compound_factor(best_under['difference'])
best_under
```

```python
max_equal_lines = best_under.groupby('game_id')['equal_lines'].transform('max')

# Filter the DataFrame to keep only the rows with the maximum value in the "equal_lines" column for each "game_id"
unders = best_under[best_under['equal_lines'] == max_equal_lines]

unders
```

```python
unders = unders.sort_values(by='equal_lines', ascending=False)
unders
```

```python
columns_to_drop = ['difference', 'difference2', 'equal_lines', 'over', 'decimal_under']
unders = unders.drop(columns_to_drop, axis=1)
unders
```

```python
column_mapping = {
    'book_name': 'book_name_under',
    'book_id': 'book_id_under',
    'total1': 'total1_under',
    'under': 'under_under',
}

# Rename only the specified columns by adding "_under" to each column name
unders.rename(columns=column_mapping, inplace=True)
unders
```

```python
column_mapping = {
    'book_name': 'book_name_over',
    'book_id': 'book_id_over',
    'total1': 'total1_over',
    'over': 'over_over',
}

# Rename only the specified columns by adding "_over" to each column name
overs.rename(columns=column_mapping, inplace=True)
overs
```

```python
best_bets = pd.merge(overs, unders, on=['GAME_DATE_EST','game_id', 'home_team', 'away_team', 'SEASON', 'PTS_home', 'FG_PCT_home', 'FT_PCT_home', 'FG3_PCT_home', 'AST_home','REB_home', 'PTS_away', 'FG_PCT_away', 'FT_PCT_away', 'FG3_PCT_away', 'AST_away', 'REB_away', 'HOME_TEAM_WINS', 'PTS_comb_actual'], how='outer')
best_bets
```
