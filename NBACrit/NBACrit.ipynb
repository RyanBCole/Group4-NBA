```python
import pandas as pd
```

```python
bet_totals = pd.read_csv(r'C:\Users\redye\Desktop\NBACrit\nba_betting_totals.csv')
games = pd.read_csv(r'C:\Users\redye\Desktop\NBACrit\games.csv')
teams = pd.read_csv(r'C:\Users\redye\Desktop\NBACrit\nba_teams_all.csv')
bet_totals
```

```python
games.rename(columns={'HOME_TEAM_ID': 'a_team_id', 'VISITOR_TEAM_ID': 'team_id', 'GAME_ID': 'game_id'}, inplace=True)
games.head()
```

```python
teams
```

```python
merged_df = pd.merge(games, teams, on='team_id')
nba_df = pd.merge(merged_df, bet_totals, on=['team_id', 'game_id', 'a_team_id'])
nba_df
```

```python
nba_df.head()
```

```python
nba_df['GAME_DATE_EST'] = pd.to_datetime(nba_df['GAME_DATE_EST'])

# Create a boolean mask to filter rows with dates between "9/30/2017" and "6/8/2018"
mask = (nba_df['GAME_DATE_EST'] >= '2017-09-30') & (nba_df['GAME_DATE_EST'] <= '2018-06-08')

# Use the mask to filter the rows and create a new DataFrame containing the desired dates
nba2017 = nba_df.loc[mask]
nba2017
```

```python
nba2017.to_csv(r'C:\Users\redye\Desktop\NBACrit\nba2017.csv', index=False)
```

```python
nba2017_copy = nba2017.copy()

```

```python
nba2017_copy.rename(columns={'a_team_id': 'home_team', 'team_id': 'away_team'}, inplace=True)
```

```python
columns_to_drop = ["GAME_STATUS_TEXT", "abbreviation", "max_year", "total2", "league_id", "min_year", "TEAM_ID_away", "TEAM_ID_home"]
nba2017_copy.drop(columns=columns_to_drop, inplace=True)
```

```python
nba2017_copy.head()
```

```python
nba2017_copy['book_id'].nunique()
```

```python
nba2017_copy.rename(columns={'price1': 'over', 'price2': 'under'}, inplace=True)
nba2017_copy
```

```python
nba2017_copy['PTS_comb_actual'] = nba2017_copy.apply(lambda row: row['PTS_home'] + row['PTS_away'], axis=1)
nba2017_copy.tail()
```

```python
nba2017_copy.to_csv(r'C:\Users\redye\Desktop\NBACrit\nba2017_clean.csv', index=False)
```

```python
best_overs = nba2017_copy.groupby(['game_id', 'total1'])['over'].idxmax()

# Drop the rows with indices that are not in the lowest_price_indices
best_over = nba2017_copy.loc[best_overs]

# Display the new DataFrame with only the rows where "price1" is the lowest for each "game_id" and "total1" combination
best_over
```

```python
best_unders = nba2017_copy.groupby(['game_id', 'total1'])['under'].idxmax()

# Drop the rows with non ideal under bets
best_under = nba2017_copy.loc[best_unders]

# Display the new DataFrame with only the best under bets for each total per game
best_under
```
