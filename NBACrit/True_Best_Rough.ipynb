```python
import pandas as pd
```

```python
best_over = pd.read_csv(r'C:\Users\redye\Desktop\NBACrit\best_over.csv')
```

```python
best_over
```

```python
first_values = best_over.groupby('game_id')['total1'].transform('first')
best_over['difference'] = best_over['total1'] - first_values

best_over
```

```python
best_over['game_id'].nunique()
```

```python
first_values
```

```python
#each half point is worth 5.1 cents so multiple diff by 10.2 so that each .5 diff becomes 5, and then add/subtract from real line
best_over['difference2'] = best_over['difference'] * 10.2
best_over
```

```python
#subtract d2 from over and make new column for comparing lines on equal footing
best_over['equal_lines'] = best_over['over'] - best_over['difference2']
best_over
```

```python
max_equal_lines = best_over.groupby('game_id')['equal_lines'].transform('max')

# Filter the DataFrame to keep only the rows with the maximum value in the "equal_lines" column for each "game_id"
overs = best_over[best_over['equal_lines'] == max_equal_lines]

overs
```

```python
overs['game_id'].nunique()
```

```python
columns_to_drop = ['difference', 'difference2', 'equal_lines', 'under']
overs = overs.drop(columns_to_drop, axis=1)
overs
```
