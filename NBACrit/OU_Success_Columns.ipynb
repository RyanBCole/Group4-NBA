```python
import pandas as pd
```

```python
best_bets = pd.read_csv(r'C:\Users\redye\Desktop\NBACrit\merged_df.csv')
```

best_bets

```python
best_bets["over_successful"] = best_bets["PTS_comb_actual"] - best_bets["total1_over"]

# Map the values to 1 or 0 based on the conditions
best_bets["over_successful"] = best_bets["over_successful"].apply(lambda x: 1 if x >= 0 else 0)
best_bets
```

```python
best_bets["under_successful"] = best_bets["total1_under"] - best_bets["PTS_comb_actual"]

# Map the values to 1 or 0 based on the conditions
best_bets["under_successful"] = best_bets["under_successful"].apply(lambda x: 1 if x >= 0 else 0)
best_bets
```
